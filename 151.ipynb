 <h3>8</h3>
                <p>Considere uma MP que possui 4K blocos de 128 células de 1 byte cada e uma memória cache do tipo associativa por conjunto que possui 64 linhas divididasem conjuntos de quatro linhas. Qual deverá ser o formato do campode endereçamento para acesso à MP e à memória cache</p>

```python
#Armazenamento da Memória
MP = pow(2,19)
#Conjunto
conjunto = 4
#Blocos
blocos = 4*pow(2,10)
#Totalde bits
Totalbits = log2(MP)
#bits do Byte
byte = log2(128)
#bits do Conjunto
conj = log2(blocos/conj)
#bits da Tag
tag = Totalbits-(byte+conj)
print("Tag: {} bits/Conjunto: {} bits/ Byte: {} bits".format(tag,conj,byte))
```
