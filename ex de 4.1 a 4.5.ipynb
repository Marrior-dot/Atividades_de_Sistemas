# QUESTÃO 4.1

<blockquote> Uma memória RAM(MP) tem um espaço máximo de endereçamento de 2K. Cada célula pode armazenar 16 bits. Qual o valor total de bits que podem ser armazenados nesta memória e qual o tamanho de cada endereço </blockquote>

```python
print("Sendo k = 2^10 e que o Barramento de endereços será igual ao logaritmo do total de endereços na base 2,temos que:")
T = pow(2,11)
BE = log2(T)
print("O barramento de endereços é igual à 11 bits")
```

```python
print("A capacidade total será resultado do produto da total de endereços pela capacidade de cada célula")
T = pow(2,11)
C = 16
Ca = T*C
print("A capacidade da total da memória é igual à 32Kbits ou {} bits".format(Ca))
```

# QUESTÃO 4.2

<blockquote> Uma memória RAM(MP) é fabricada com a possibilidade de armazenar no máxmio 256Kbits. Cada célula pode armazenar 8 bits. Qual é o tamanho de cada endereço e qual é o total de células que podem ser usadas naquela RAM. </blockquote>

```python
print("256K = 2^8*2^10")
T = pow(2,15)
BE = log2(T)
print("O tamanho do barramento de endereços é igual a {}".format(BE))
print("O total de células é igual à 32K ou {}".format(T))
```

# QUESTÃO 4.3

<blockquote>Um computador, cuja memória RAM(MP) tem uma capacidade máxima de armazenamento de 2K palavras de 16 bits cada, possui um REM e um RDM. Qual o tamanho desses registradores, qual o valor do maior endereço dessa MP e qual a quantidade total de bits que podem ser armazenados </blockquote>

```python
print("2K = 2*2^10")
C = 16
T = pow(2,11)
BE = log2(T)
BD = C
Ca = T*C
print("O REM terá tamanho igual a 2K ou {}, já o RDM terá tamanho igual a {}".format(BE,BD))
print("A capacidade total será de 32Kbits ou {} bits".format(Ca))
```

# QUESTÃO 4.4

<blockquote>Um RDM possui capacidade de armazenar 32 bits e um REM com capacidade de armazenar 24 bits.Sabendo que em cada acesso são lidas 2 células da memória RAM(MP) e que o barramneto de Dados tem tamanho igual à palavra.Pergunta-se </blockquote>
