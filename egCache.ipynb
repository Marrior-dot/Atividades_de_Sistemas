### 5.5 
<p>Considere um sistema de computação com uma memória de 32KB de capacidade, constituída de linhas com 8 bytes de largura.Qual a quantidade de bits necessária para essa memória cache?.</p>

```python
print("A quantidade de bits necessária para a memória é igual a soma dos dados da cache mais o total de bits do campo de blocos.") 
K = pow(2,10) #1024bits
B = 8 #valor do byte
TD = 32*K*B #Capacidade da memória cache
```

<p>O total de bits do campo bloco é igual ao produto da quantidade de linhas pela largura do bloco</p>
<blockquote><strong>L(quantidade de linhas)</strong>=32KB/8B = 2^5*2^10*2^-3 = 4KB ou 2<sup>12</sup> </blockquote>
<blockquote><strong>b(largura do campo bloco)</strong>=16MB/8B = 2<sup>4</sup>*2<sup>20</sup>*2<sup>-3</sup> = 2<sup>21</sup>
    <p>O bloco possui de largura 21 bits</p>    
</blockquote>

```python
K = pow(2,10)
Tb = 4*K*21 #Total de bits do campo bloco
T = TD + Tb
print("A memória precisa de {} bits".format(T))
```

### 5.9
<p>Considere uma MP com 64MB de capacidade associada a uma memória cache que funciona com mapeamento assosciatvo por conjunto de 4 e que possui 32KB, com linhas de largura de 16 bytes.Determine o formato de endereço para ser interpreto pelo sistema de controle de cache.</p>

<blockquote>
    MP = 2<sup>26</sup>
    <br>
    largura do bloco = 2<sup>4</sup>
    <br>  
    
</blockquote> 

```python

bl = log2(pow(2,4))
print("{}bytes por linha".format(bl))
```

<blockquote>
    <br>
    Capacidade da cache = 32KB
    <br>
    Blocos da MP = 2<sup>26</sup>/2<sup>4</sup> = 2<sup>22</sup>
    <br>
    <b>2<sup>22</sup> ou 4M blocos</b>
    <br>
    Linhas da cache = 2<sup>15</sup>/2<sup>4</sup> =2*2<sup>10</sup>
    <b>2<sup>11</sup> ou 2K linhas</b>
    <br>
    Conjuntos = Linhas/4 = 2<sup>11</sup>/2<sup>2</sup> =  2<sup>9</sup>
    <br>
    <b>2<sup>9</sup>conjuntos</b>
    

```python
C = log2(pow(2,9))
print("{}bits para o conjunto".format(C))
```

<blockquote>Blocos por conjunto = 4M/2<sup>9</sup> = 2<sup>22</sup>/2<sup>9</sup> = 2<sup>13</sup> 
    <br>
    <b>2<sup>13</sup>ou 8K Blocos por conjunto</b>
    <br>
    <b>Campo Tag = endereços da mp - (conjuntos+bytes)</b></blockquote>

```python
tag = 26-(9+4)
print("{} bits para a tag".format(tag))
```

 Formato = 
    <b>tag</b>/<b>conjunto</b>/<b>byte</b>

<blockquote> <b>13</b>/<b>9</b>/<b>4</b> </blockquote>

<p>Seja uma MP constituída de blocos com largura de 32 bytes, associada a uma cache com 64KB; a cache usa mapeamento por conjunto de 4. Em dado instante o processador realiza um acesso colocando o seguinte endereço(expresso em algarismos hexdecimais); 3FC92B6. Determine qual deverá ser o valo binário do conjunto que será localizado pelo sistema de controle da cache.</p>

<blockquote>O endereço da MP possui 7 hexadecimais</blockquote>

```python
h = 4
print("{} bits para o endereço da MP".format(7*h))
```

<blockquote>32bytes por bloco/linha</blockquote>


```python
c = log2(32)
print("{} bits para o campo byte".format(c))
```

<blockquote>Quantidade de linhas = Total da cache/Largura da linha</blockquote>

```python
M = 64 #Capacidade da cache em KB
Lb = 32 #Largura da linha em B
L = M/Lb 
print("{}K linhas".format(L))
```

<blockquote>Conjuntos  = Linhas/4
    <br>
    2K/4 = 2<sup>11</sup>/2<sup>2</sup> = 2<sup>9</sup>
    <br>
    <b>9 bits para o campo de conjuntos</b>
</blockquote>


<blockquote>Tag = Endereço da MP-(conjuntos+byte)</blockquote>

```python
MP = 28
conj = 9
byte = 5
tag = MP - (conj+byte)
print("{} bits para a tag".format(tag))
```

<blockquote>Formato =
