<p>Um computador possui uma memória principal com capacidade para armazenar palavras de 16 bits em cada uma das células, e o seu barramento de endereços tem 12 bits de tamanho. Sabendo-se que em cada célula pode-se armazenar o valor exato de uma palavra, quantos bytes poderão ser armazenados nessa memória.</p>

```python
BE = pow(2,12) #Número de Células
C = 16 #Bits por Célula
Ca = (BE*C)/8 #Capacidade da Memória em bytes
print("A capacidade da memória em bytes é 8KB ou {}bytes".format(Ca))
```

### 6
<p>Qual é a função do Registrador de Memória(RDM) e do Registrador de Endereços (REM)?</p>

#### REM:
<p>Armazena temporariamente os bits de endereço antes de uma operação de leitura ou escrita, permite armazenar a mesma quantidade de bits do barramento de endereço</p>

#### RDM:
<p>Armazena os dados temporariamente, numa transferência do registrador do processador para a MP ou vice-versa, permite armazenar a mesma quantidade de bits do barramento de dados </p>

### 7
<p>Descreva os barramentos que interligam o processador e a MP, indicando função e direção do fluxo de cada um.</p>

#### BD(Barramento de Dados)
<p>Responsável por transferir a informação que fica armazneda temporariamente no RDM durante as operações de leitura e escrita é bidirecional, pois a informação pode ir da MP para o processador quanto do processador para a MP.</p>

#### BE(Barramento de Endereços)
<p>Responsável por transferir os bits de endereço para endereçar uma célula na MP, o qual fica armazenado temporariamente no REM durante as operações de leitura e escrita, é unidirecional, pois os bits de endereço só vão para da processador para a MP.</p>
    
#### BC(Barramento de Controle)
<p>Responsável por mandar o comando de leitura, escrita, ou wait. Para a leitura, temos que o processador pode enviar sinais de operação de leitura ou de escrita para a MP e a MP pode enviar sinais de espera para a realização de outras operações para o processador. </p>

### 15
<p>Compara uma memória principal e uma memória cache em termos de tempo de acesso, capacidade e temporariedade de armazenamento de dados</p>

<blockquote><strong>Hierarquia de Memórias:</strong> 
    <ol>
        <li>Registradores</li>
        <li>Memória Cache</li>
        <li>Memória Principal</li>
        <li>Memória Secundária</li>
    </ol>
</blockquote>

<p><strong>Tempo de Acesso:</strong>Como é possível ver na hierarquia de memórias
temos que a memória cache está acima da MP, portanto o tempo de acesso é menor que o da MP, já que a memória cache apenas armazena dados por tempo suficiente para que eles sejam transferidos entre a MP e o processador
</p>
<p><strong>Capacidade:</strong>Na hierarquia a memória cache está em segundo lugar e a MP em terceiro lugar, portanto temos que a memória possui capacidade total maior, na ordem dos GB(4GB,8GB, 16GB, NGB....), enquanto que a memória cache possui capacidade total menor na ordem dos MB(4MB, 8MB,16MB) isso se deve às funções de cada memória, sendo a da MP de armazenar informações necessárias para armazenar as instruções necessárias para que certos programas funcionam, enquanto que a memória cache apenas transfere essas informações para o processaodr. 
</p>
<p><strong>Temporariedade:</strong>Como a memória cache vem em segundo lugar na hierarquia, ela armazena dados por menos tempo que a MP,que vem em terceiro lugar, porém ambas são de característica transitória, ou seja armazenam os dados por tempo necessário apenas para realizar as suas funções.</p>
